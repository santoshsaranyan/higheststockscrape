```python
import yfinance as yf
import pandas as pd
```

```python
df=pd.read_csv("lab_assignment.csv")
df
```

```python
new = {"Ticker":df["TICKER"],"Year":df["YEAR"]}
new_df=pd.DataFrame(data=new)
new_df=new_df.drop_duplicates()
new_df
```

```python

for i in range(0,len(new_df.index)):
    if ("." in new_df.iloc[i,0]):
        print(new_df.iloc[i,0])
        new_df.iloc[i,0]= new_df.iloc[i,0].replace(".","-")
        print(new_df.iloc[i,0])
```

```python
val=[]
for i in range(0,len(new_df.index)):
    tckr=yf.Ticker(new_df.iloc[i,0])
    year=str(new_df.iloc[i,1])
    d1=year+"-01-01"
    d2=year+"-12-31"
    inf=tckr.history(start=d1, end=d2)
    print(i)
    if(inf["High"].empty==True):
        val=val+[-1]
    else:
        val=val+[max(inf["High"])]
val
```

```python
dd={"TICKER":new_df["Ticker"],"YEAR":new_df["Year"],"STOCK":val}
final_df=pd.DataFrame(data=dd)
final_df.to_csv("Highest_stocks.csv")
```
