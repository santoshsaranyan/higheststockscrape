```python
import yfinance as yf
import pandas as pd
```

```python
df=pd.read_csv("lab_assignment.csv") #Reading in the data with the tickers of the companies
df
```

```python
new = {"Ticker":df["TICKER"],"Year":df["YEAR"]} # Getting the Tickers and the year
new_df=pd.DataFrame(data=new)
new_df=new_df.drop_duplicates() # Getting the unique ticker-year combination
new_df
```

```python
# To make the ticker conform with the format of yfinance's tickers
for i in range(0,len(new_df.index)):
    if ("." in new_df.iloc[i,0]):
        print(new_df.iloc[i,0])
        new_df.iloc[i,0]= new_df.iloc[i,0].replace(".","-")
        print(new_df.iloc[i,0])
```

```python
val=[]
for i in range(0,len(new_df.index)):
    tckr=yf.Ticker(new_df.iloc[i,0])
    year=str(new_df.iloc[i,1])
    d1=year+"-01-01" # Setting a start date to get stock data from
    d2=year+"-12-31" # Setting an end date (end of the year)
    inf=tckr.history(start=d1, end=d2) # Getting the stock history for the ticker for the given year
    print(i)
    if(inf["High"].empty==True):
        val=val+[-1]
    else:
        val=val+[max(inf["High"])] # Getting the highest stock price
#val
```

```python
dd={"TICKER":new_df["Ticker"],"YEAR":new_df["Year"],"STOCK":val} #Appending the highest stock value to corresponding ticker-year
final_df=pd.DataFrame(data=dd)
final_df.to_csv("Highest_stocks.csv")
```
